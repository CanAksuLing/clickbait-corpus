```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import matplotlib
import pandas as pd
```

```python
df = pd.read_csv('survey.csv')
```

```python
df.publication_date = df.publication_date.str[:10]
date = df.publication_date.value_counts()
```

```python
matplotlib.style.use('./seaborn-v0_8-ticks.mplstyle')
plt.bar(date.index, date.values)
plt.xlabel('Publication Date (YYYY-MM-DD)')
plt.ylabel('Frequency')
plt.savefig('assets/article-date-plot.svg')
```
